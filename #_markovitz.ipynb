```python
from cvxpy import *
import numpy as np
# from posdef import nearestPD

def MarkowitzOpt(mean, Sigma, min_return = 0.01	):
    n = mean.size                        # Number of assets (number of stocks + interest rate)
    
    mu = mean#.values                    # Mean returns of n assets
        
    counter = 0
    

    
    w = Variable(n)                         # Portfolio allocation vector
    ret = mu.T@	w
    risk = quad_form(w, Sigma)        # quadratic form 
    min_ret = Parameter(nonneg=True)
    min_ret.value = min_return
    prob = Problem(Minimize(risk),          # Restricting to long-only portfolio
                   [ret >= min_ret,
                   sum(w) == 1
                    #,w >= 0
                    ])
    prob.solve()
    return w.value
```

```python

A = Sigma[0]
b =  A @ mu
x0 = np.random.randn(10)
```

```python
def subgrad_const_stepsize(f, subgrad, x0, alpha, maxiter):
  conv = [x0.copy()]
  x = x0.copy()
  for i in range(maxiter):
    g = subgrad(x)
    x = x - alpha * g
    conv.append(x.copy())
  return x, conv

def subgrad_decay_stepsize(f, subgrad, x0, maxiter):
  conv = [x0.copy()]
  x = x0.copy()
  for i in range(maxiter):
    g = subgrad(x)
    x = x - g / (i + 1)
    conv.append(x.copy())
  return x, conv

def subgrad_scaled_stepsize(f, subgrad, x0, alpha, maxiter):
  conv = [x0.copy()]
  x = x0.copy()
  for i in range(maxiter):
    g = subgrad(x)
    x = x - alpha * g / np.linalg.norm(g)
    conv.append(x.copy())
  return x, conv
```

```python
f = lambda x: np.linalg.norm(A @ x - b, 1)                        # Не разобрался с целевой функцей?!
subgrad = lambda x: A.T @ np.sign(A @ x - b)                               # sign - это любо плюс либо минус знак  // но в итоге получаем вектор???
alpha = 1e-2
maxiter = 80000
```

```python
_, conv_constsize = subgrad_const_stepsize(f, subgrad, x0, alpha, maxiter)
_, conv_decay_size = subgrad_decay_stepsize(f, subgrad, x0, maxiter)
_, conv_scaled_size = subgrad_scaled_stepsize(f, subgrad, x0, alpha, maxiter)
```
